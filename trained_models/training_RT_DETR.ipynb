# **RT-DETR (Real-Time Detection Transformer)**

```python
from ultralytics import YOLO

DATA_YAML = "/content/Wild-Animals-1/data.yaml"

with open(DATA_YAML, "r") as f:
    print(f.read())
```

```python
from ultralytics import YOLO
import time

model = YOLO("rtdetr-l.pt")
```

```python
start_time = time.time()

model.train(
    data="/content/Wild-Animals-1/data.yaml",
    epochs=5,
    imgsz=640,
    batch=8,
    device=0,
    workers=4,
    optimizer="AdamW",
    lr0=1e-4,
    project="RTDETR_Animals",
    name="probe_5epochs"
)

end_time = time.time()
print(f"⏱️ Time for 5 epochs: {(end_time - start_time)/60:.2f} minutes")
```

```python
model.val(
    data="/content/Wild-Animals-1/data.yaml",
    split="val",
    imgsz=640,
    device=0
)
```

```python
model.val(
    data="/content/Wild-Animals-1/data.yaml",
    split="test",
    imgsz=640,
    device=0
)
```
