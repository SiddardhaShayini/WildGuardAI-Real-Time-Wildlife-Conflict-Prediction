```python
from ultralytics import YOLO

# Load the augmented model
model_augmented = YOLO('best_augmented.pt')

print(f"✅ Augmented Model Task: {model_augmented.task}")
print(f"✅ Number of Classes: {len(model_augmented.names)}")
```

```python
# Run validation on the augmented model
metrics_augmented = model_augmented.val(
    data="/content/Wild-Animals-1/data.yaml",
    imgsz=960,
    device=0,
    augment=True, # Ensure TTA is applied during validation if it was trained with it
    conf=0.001,
    iou=0.7
)
```

### Performance Metrics for `best_augmented.pt`

```python
print(f"Precision (P): {metrics_augmented.box.p}")
print(f"Recall (R): {metrics_augmented.box.r}")
print(f"mAP50: {metrics_augmented.box.map50}")
print(f"mAP50-95: {metrics_augmented.box.map}")

# Note: YOLO models typically report precision, recall, and mAP (mean Average Precision) scores.
# Accuracy and F1-score are not directly reported in the same way for object detection models,
# but can be derived from the precision-recall curves or confusion matrices if needed.
```

### Validating with a Tuned IoU Threshold (`iou=0.6`)

```python
from ultralytics import YOLO

# Load the augmented model
model_augmented = YOLO('best_augmented.pt')

# Run validation on the augmented model with a tuned iou threshold
# We'll use imgsz=960 again, as 1280 didn't improve performance.
metrics_tuned_iou = model_augmented.val(
    data="/content/Wild-Animals-1/data.yaml",
    imgsz=960,
    device=0,
    augment=True, # Keep TTA on
    conf=0.001,
    iou=0.6 # Tuned IoU threshold (default was 0.7)
)
```

### Performance Metrics with Tuned IoU

```python
print(f"Precision (P): {metrics_tuned_iou.box.p}")
print(f"Recall (R): {metrics_tuned_iou.box.r}")
print(f"mAP50: {metrics_tuned_iou.box.map50}")
print(f"mAP50-95: {metrics_tuned_iou.box.map}")
```

```python
# Save the model (this will overwrite 'best_augmented.pt' if it already exists)
model_augmented.save('best_augmented.pt')
print("Model 'best_augmented.pt' saved successfully.")
```
